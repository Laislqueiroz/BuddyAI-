```python
#importação das bibliotecas

import matplotlib.pyplot as plt
import numpy as np # operações matematicas complexas e manipulação de arrays
from sklearn.linear_model import LinearRegression # para criar modelos de regressão linear
from sklearn.metrics import mean_squared_error, r2_score # para calcular metricas de avaliação do modelo

# Dados fornecidos pelo usuario

gastos_publi = [1.3, 2.4, 4.2, 5.6, 7.7, 8.9, 9.1, 4.6, 1.9, 2.8]
licencas_vend = [50, 12, 105, 90, 43, 76, 28, 88, 53, 63]

#conversão das listas em array numpy.

x = np.array(gastos_publi).reshape(-1, 1) * 1000 # gastos em publi
y = np.array(licencas_vend)

# criação do modelo de regressão usando a bibli scikit_learn

modelo = LinearRegression()
modelo.fit(x, y)

y_pred = modelo.predict(x) #predição dos valores de y com base nos valores x
mse = mean_squared_error(y, y_pred) # criando erro quadratico medio
r2 = r2_score(y, y_pred) # calculo do coeficiente de determinação

#Geração e configuração do grafico para visualizar os dados e a linha de regressão
plt.figure(figsize=(10, 6)) # definição do tamanho da figura do grafico
plt.scatter(x, y, color='black', label='Dados Reais') # Grafico de disperssão para os dados reais
plt.plot(x, y_pred, color='blue', linewidth=2, label='linha de regressão') # Linha de regressão

#Adição de titulos e rotulos aos eixos do grafico

plt.title('Regressão Linear - Vendas e Gastos')
plt.xlabel('Gastos com Publi')
plt.ylabel('Licenças Vendas')
plt.legend() #inclusão de legenda
plt.show() #visualização do grafico


#Uso do modelo treinado para prever o numero de licenças vendidas com um gasto de 8000 em publis

previsao = modelo.predict(np.array([[8000]]))

# os resultados das metricas de avaliação e as previsões são apresentadas

mse, r2, previsao
```
