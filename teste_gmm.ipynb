```python
import numpy as np
import matplotlib.pyplot as plt
from sklearn.mixture import GaussianMixture

# Definir uma semente aleatoria para garantir a reprodutibilidade dos resultados

np.random.seed


# Criação em dois conjuntos de daods ficticios cada um com distribuição gauseana

data1 = np.random.normal(loc =-5, scale=1, size=300) # 300 pontos com media -5 e desvio padrão 1

data2 = np.random.normal(loc =5, scale=1, size=700) # 700 pontos com media 5 e desvio padrão 1

# Combinar os dois dados para formar um unico conjunto

data = np.concatenate([data1, data2])

# Visualização do conunto de dados originais em um grafico.

plt.hist(data, bins=30, density=True)
plt.title("Originals")
plt.show()

# Criando e treinando o modelo de mistura Gauseana de dois componetes

gmm = GaussianMixture(n_components=2)

# Ajusta modelo de Dados

gmm.fit(data.reshape(-1,1))

# Usando o treinamento para gerar 1000 novos pontos de dados

samples = gmm.sample(1000) [0]
plt.hist(samples, bins=30, density=True, alpha = .5, color = 'purple', label= 'BTS JIN')
plt.hist(data, bins=30, density=True, alpha = .5, color = 'yellow', label= 'BTS Hope')
plt.title('BTS Mix')
plt.legend()
plt.show()

```
